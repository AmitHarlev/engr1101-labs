## Part 1: What is a Maximum Bipartite Matching?

In this part of the lab, we will look at a small example to see how a maximum bipartite matching can be found using max flow. We will start by reviewing the definition of a maximum bipartite matching.

A **bipartite matching** is a set of edges in a bipartite graph such that no two edges share an endpoint. In other words, each chosen edge pairs one node from the left side with one node from the right side, and each node can appear in at most one chosen pair.

A **maximum bipartite matching** is a bipartite matching with as many edges as possible. Equivalently, it matches as many left-right pairs as the graph allows without reusing any node.

Just as in the max flow lab, we can use NetworkX to quickly solve the max flow problem. Below is an example in which we:
 - transform the bipartite graph into the corresponding flow problem,
 - solve the flow problem using NetworkX,
 - recover the maximum matching from the maximum flow.

Read the code to get a sense for what it is doing. In the next part we will use NetworkX to solve a much larger bipartite matching problem with real data.

```python
# Build the max-flow network for the toy bipartite matching problem.
G = nx.DiGraph()
G.add_node("s")
G.add_node("t")

for left in left_nodes:
    G.add_edge("s", left, capacity=1)

for left, right in compatibility_edges:
    G.add_edge(left, right, capacity=1)

for right in right_nodes:
    G.add_edge(right, "t", capacity=1)
```

```python
# Solve the max flow problem and recover the maximum matching.
flow_value, flow_dict = nx.maximum_flow(G, "s", "t")
matching = [(left, right) for left, right in compatibility_edges if flow_dict[left][right] == 1]

matching_indices = [
    (left_nodes.index(left), len(left_nodes) + right_nodes.index(right))
    for left, right in matching
]

print("Maximum flow value:", flow_value)
print("Maximum matching:")
for edge in matching_indices:
    print(edge)
```

Run the next cell to run the greedy algorithm and see how it does.

```python
## Match taxis to passengers using a greedy algorithm

# Get all edges in the graph
all_taxi_to_passenger_edges = build_all_taxi_to_pickup_arcs(available_taxis, pickup_requests, nodes_df)

# Lists/sets to keep track of which edges/taxis/passengers we've already selected
greedy_matching = []
used_taxis = set()
used_pickup_requests = set()

# Sort the edges by travel time
def travel_time(edge):
    taxi, pickup_request, cost = edge
    return cost
all_taxi_to_passenger_edges = sorted(all_taxi_to_passenger_edges, key=travel_time)

# Since edges are sorted by travel time, add an edge to the matching if we haven't already used that taxi or passenger
for taxi, pickup_request, cost in all_taxi_to_passenger_edges:
    if taxi not in used_taxis and pickup_request not in used_pickup_requests:
        greedy_matching.append((taxi, pickup_request, cost))
        used_taxis.add(taxi)
        used_pickup_requests.add(pickup_request)

# Get the line between each matched taxi and passenger for visualization
all_pairs_greedy_paths = [
    [(taxi[:4], pickup_request[:4], False)]
    for taxi, pickup_request, cost in greedy_matching
]

# Find how many passengers got a ride under the time threshold
passengers_with_ride_under_threshold = len([
    pickup_request
    for taxi, pickup_request, cost in greedy_matching
    if cost / 1000 <= pickup_time_threshold
])

print(f"Passengers with a ride under pickup time threshold: {passengers_with_ride_under_threshold}")

show(plot_taxi_route(all_pairs_greedy_paths, nodes_df, "Greedy Matching Without Travel-Time Cutoff"))
```

```python
edges_under_threshold = build_taxi_assignment_arcs(
    available_taxis,
    pickup_requests,
    nodes_df,
    pickup_time_threshold,
)

print(f"There are {len(edges_under_threshold)} taxi-to-pickup request edges under the threshold.")
```

Why do this? In many bipartite graphs, finding a _pretty good_ matching is not that hard (for example, by randomly choosing edges). So we want to see how much of an improvement we are getting by going through all the trouble of learning and using the maximum matching algorithm.

Since the above algorithm is random, it can get many different results if you re-run it multiple times. Let's run it 100 times to get a sense for the average performance.

```python
# Run the random matching process 100 times to measure average performance.
random_matching_sizes = []

for trial in range(100):
    random_matching = random_taxi_matching(edges_under_threshold)

    # Store only the size of the matching
    random_matching_sizes.append(len(random_matching))

average_random_matching = sum(random_matching_sizes) / len(random_matching_sizes)

print(f"Average random matching size over 100 runs: {average_random_matching:.2f}")
print(f"Smallest random matching: {min(random_matching_sizes)}")
print(f"Largest random matching: {max(random_matching_sizes)}")
```

We will now use NetworkX to solve for the maximum matching.

```python
# Uses networkx to solve the bipartite matching problem
def find_optimal_matching():
    B = nx.Graph()
    B.add_nodes_from(available_taxis, bipartite=0)
    B.add_nodes_from(pickup_requests, bipartite=1)
    B.add_edges_from([edge[:2] for edge in edges_under_threshold])
        
    return nx.bipartite.maximum_matching(B, available_taxis)
    
match = find_optimal_matching()
```

```python
matching_size = int(len(match) / 2)
print(f"Maximum matching size: {matching_size}")
```

Let's also visualize what this matching of taxis to passengers looks like on the map.

```python
# Build paths for visualizing the maximum matching on the map.
max_match_paths = []

for taxi, pickup_request in match.items():
    if taxi not in available_taxis:
        continue

    max_match_paths.append([(taxi[:4], pickup_request[:4], False)])
```

```python
show(plot_taxi_route(max_match_paths, nodes_df, 'Maximum Matching'))
```
